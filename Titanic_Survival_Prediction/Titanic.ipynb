```python
titanic_data.replace({'Sex':{'male':0,'female':1}, 'Embarked':{'S':0,'C':1,'Q':2}}, inplace=True)
```

```python
titanic_data.head()
```

```python
titanic_data['Age'] = titanic_data['Age'].fillna(titanic_data['Age'].mean())

titanic_data['Embarked'] = titanic_data['Embarked'].fillna(
    titanic_data['Embarked'].mode()[0]
)
```

```python
print(titanic_data[['Age', 'Embarked']].isnull().sum())
```

```python
titanic_data['Sex'] = titanic_data['Sex'].astype(int)
titanic_data['Embarked'] = titanic_data['Embarked'].astype(int)
```

```python
X = titanic_data.drop(columns=['PassengerId','Name','Ticket','Survived'])
Y = titanic_data['Survived']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=2
)
```

```python
print(X_train.isnull().sum())
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)

model.fit(X_train, Y_train)
```

```python
print(X_train.isnull().sum())
```

```python
X_train_prediction = model.predict(X_train)
```

```python
print(X_train_prediction)
```

```python
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy score of training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
```

```python
print(X_test_prediction)
```

```python
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy score of test data : ', test_data_accuracy)
```
